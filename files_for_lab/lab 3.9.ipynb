## Load the dataset and explore the variables.

```python
data=pd.read_csv('customer_churn.csv')
data.head(50)
```

## We will try to predict variable Churn using a logistic regression on variables tenure, SeniorCitizen,MonthlyCharges.


```python
data1 = pd.DataFrame([data.SeniorCitizen , data.tenure, data.MonthlyCharges, data.Churn]).T
data1
```

```python
data1['MonthlyCharges'] = data1['MonthlyCharges'].astype(float)
data1['tenure'] = data1['tenure'].astype(int)
data1['SeniorCitizen'] = data1['SeniorCitizen'].astype(int)
```

```python
data1['Churn'].value_counts()
```

```python
data1.dtypes
```

```python
data1.isna().sum()
```

## Split the Dataset into X ('tenure', 'SeniorCitizen', 'MonthlyCharges') and y ('Churn')

```python
# X/y split

X = data1[['SeniorCitizen','tenure','MonthlyCharges']]
y = data1['Churn']
```

```python
correlations_matrix = X.corr()
sns.heatmap(correlations_matrix, annot=True)
plt.show()
```

```python
data1
```

```python
data1.dtypes
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0) 
```

## Build the logistic regression model.

```python
# all features are numeric, so no need to split into _num and _cat
transformer = StandardScaler().fit(X_train)
X_train_normalized = pd.DataFrame(transformer.transform(X_train),columns=X.columns)
# because this is the only tranformation we apply to the numerics, 
# we can immediately transform the X_test as well
X_test_normalized = pd.DataFrame(transformer.transform(X_test),columns=X.columns)
X_train_normalized.head()
X_test_normalized.head()
```

```python
LR = LogisticRegression(random_state=0, solver='lbfgs')
LR.fit(X_train_normalized, y_train)
```

```python
#make predictions using X_test_transfomed : model.predict -> predictions
```

## Evaluate the model.

```python
LR.score(X_test_normalized, y_test) # a.k.a Accuracy
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score 
from sklearn.metrics import accuracy_score

pred = LR.predict(X_test_normalized)

print("accuracy: ", accuracy_score(y_test,pred))
print("precision: ", precision_score(y_test,pred, pos_label='Yes'))
print("recall: ",recall_score(y_test,pred, pos_label='Yes'))
print("f1: ",f1_score(y_test,pred, pos_label='Yes'))
```

```python
from sklearn.metrics import confusion_matrix



confusion_matrix(y_test,pred)

#      TN,FP,
#      FN,TP
```

## Even a simple model will give us more than 70% accuracy. Why?

```python
# there is a clear majority of churn=no. The dataset is imbalanced
```

## Synthetic Minority Oversampling TEchnique (SMOTE) is an over sampling technique based on nearest neighbors that adds new points between existing points. Apply imblearn.over_sampling.SMOTE to the dataset. Build and evaluate the logistic regression model. Is it there any improvement?

```python
from imblearn.over_sampling import SMOTE
```

```python
sm = SMOTE(random_state=100, k_neighbors=3)
X_train_SMOTE,y_train_SMOTE = sm.fit_resample(X_train_normalized,y_train)
```

```python
X_train_SMOTE.shape
```

```python
LR2 = LogisticRegression(random_state=0, solver='lbfgs')
LR2.fit(X_train_SMOTE, y_train_SMOTE)
```

```python
LR2.score(X_test_normalized, y_test) 
```

```python

pred = LR2.predict(X_test_normalized)

print("precision: ",precision_score(y_test,pred, pos_label='Yes'))
print("recall: ",recall_score(y_test,pred, pos_label='Yes'))
print("f1: ",f1_score(y_test,pred, pos_label='Yes'))
```

```python
confusion_matrix(y_test,pred)
```

```python
#accuracy, precision and f1 went down, only recall went up. This is not an improvement.
```
